```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nb15_sample_path = 'nb_sample_almost_all_feat.csv'
```

```python
nb15_sample_dataset = pd.read_csv(nb15_sample_path)
```

```python
# Generate KDE plot for 'column_name'
ax = nb15_sample_dataset['dur'].plot.kde()

# Customize plot
ax.set_title('KDE Plot for `dur`')
ax.set_xlabel('`dur`')
ax.set_ylabel('Density')

# Show plot
plt.show()
```
