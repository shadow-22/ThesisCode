```python
folder = 'C:/Users/katsa/OneDrive/Jupyter_files/pytorch/gan_nb15_all_features/version1.1/test_eval2/'
```

```python
output_lst = glob.glob(folder + '*')
print(output_lst)
```

```python
acc_lst = []
for output_file in output_lst:
    with open(output_file, "r") as file:
        first_line = file.readline().strip()
        substring = first_line.split(":", 1)[-1].strip()
        accuracy = float(substring)
        acc_lst.append(accuracy)
```

```python
print(len(acc_lst))
```

```python
import matplotlib.pyplot as plt

data = acc_lst

x = range(len(data))
y = data

plt.scatter(x, y)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Scatter Plot')
plt.savefig('epoch_vs_accuracy_plot.png')
plt.show()
```
