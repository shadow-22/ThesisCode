```python
import pandas as pd
import numpy as np
```

```python
train_dataset_path = 'C:/Users/katsa/Documents/cicids_sample.csv'
CICDataset_train = pd.read_csv(train_dataset_path)
X_train, y_train = CICDataset_train.iloc[:, :-1], CICDataset_train['Label']
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
print(y_train.value_counts())
print(y_train.value_counts().plot.pie())
```

```python
for label in y_train:
    if (label != 'BENIGN' and label != 'ATTACK'):
        y_train.replace(label, 'ATTACK', inplace=True)
        
print(y_train.value_counts())
print(y_train.value_counts().plot.pie())
```

```python
df_result = pd.concat([X_train, y_train], axis=1)
df_result
```

```python
df_result.to_csv("cic_train_sample_binary.csv", index=False)
```
