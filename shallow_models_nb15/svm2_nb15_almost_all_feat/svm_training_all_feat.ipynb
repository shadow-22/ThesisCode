```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(
    SimpleImputer(), StandardScaler(), SVC(kernel='rbf', gamma=0.001, C=1)
)
```

```python
nb_train_dataset_path = 'C:/Users/katsa/OneDrive/Jupyter_files/shallow_models_cic_nb15/nb_all_feat_train_dataset.csv'
nb_train_dataset = pd.read_csv(nb_train_dataset_path)
print(len(nb_train_dataset))
nb_train_dataset = nb_train_dataset.sample(frac=1/6)
print(len(nb_train_dataset))
nb_train_dataset.head()
```

```python
nb_test_dataset_path = 'C:/Users/katsa/OneDrive/Jupyter_files/shallow_models_cic_nb15/nb_all_feat_test_dataset.csv'
nb_test_dataset = pd.read_csv(nb_test_dataset_path)
print(len(nb_test_dataset))
#nb_test_dataset = nb_test_dataset.sample(frac=1/6)
print(len(nb_test_dataset))
nb_test_dataset.head()
```

```python
X_train = nb_train_dataset.drop('Label', axis=1)
y_train = nb_train_dataset['Label']

X_test = nb_test_dataset.drop('Label', axis=1)
y_test = nb_test_dataset['Label']
```

```python
pipe.fit(X_train.to_numpy(), y_train.to_numpy())
predictions = pipe.predict(X_test.to_numpy())
print(metrics.accuracy_score(y_test.to_numpy(), predictions))
print(metrics.classification_report(y_test.to_numpy(), predictions))
```

```python
import joblib
joblib.dump(pipe, "SVMmodel")
joblib.dump(pipe[1], "scaler")
```
