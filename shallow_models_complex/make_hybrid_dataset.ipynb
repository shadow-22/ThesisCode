```python
# test datasets
nb_sample_12_feat_test_dataset_path = 'C:/Users/katsa/OneDrive/Jupyter_files/shallow_models_cic_nb15/nb_12_feat_test_dataset.csv'
cic_sample_all_feat_test_dataset_path = 'C:/Users/katsa/OneDrive/Jupyter_files/shallow_models_online/cic_train_sample_binary.csv'
```

```python
# feature selection for CICIDS-2017 dataset

cic_feat_selection_vector = ['Destination Port', 'Flow Duration', 'Total Fwd Packets', 
                              'Total Backward Packets', 'Total Length of Fwd Packets', 
                              'Total Length of Bwd Packets', 'Fwd Packet Length Mean', 
                              'Bwd Packet Length Mean', 'Flow Bytes/s', 'Flow Packets/s', 
                              'Fwd Packets/s', 'Bwd Packets/s', 'Label']
```

```python
nb_12_train_df = pd.read_csv(nb_sample_12_feat_train_dataset_path)
nb_12_test_df = pd.read_csv(nb_sample_12_feat_test_dataset_path)

cic_full_train_df = pd.read_csv(cic_sample_all_feat_train_dataset_path)
cic_full_test_df = pd.read_csv(cic_sample_all_feat_test_dataset_path)
```

```python
# CICIDS-2017 Feature Selection

cic_selected_train_df = cic_full_train_df[cic_feat_selection_vector]
cic_selected_test_df = cic_full_test_df[cic_feat_selection_vector]
```

```python
# check relative size
dataset_ratio = len(cic_selected_train_df) / len(nb_12_train_df)
print(dataset_ratio)
```

```python
nb_12_train_df_reduced = nb_12_train_df.sample(frac=0.34, random_state=42)
dataset_ratio2 = len(cic_selected_train_df) / len(nb_12_train_df_reduced)
print(dataset_ratio2)
```

```python
# merging the two dataframes requires they have the same column names
new_cols = cic_feat_selection_vector
old_cols = nb_12_train_df_reduced.columns
nb_12_train_df_reduced_renamed = nb_12_train_df_reduced.rename(columns=dict(zip(old_cols, new_cols)))
nb_12_train_df_reduced_renamed.head()
```

```python
merged_train_df = pd.concat([cic_selected_train_df, nb_12_train_df_reduced_renamed], axis=0)
```
